## 2017

```python
#Fill in the name of file and join it with path

filename2017Q1="/2017Q1-capitalbikeshare-tripdata.csv"
filename2017Q2="/2017Q2-capitalbikeshare-tripdata.csv"
filename2017Q3="/2017Q3-capitalbikeshare-tripdata.csv"
filename2017Q4="/2017Q4-capitalbikeshare-tripdata.csv"


traindata2017Q1=path+filename2017Q1
traindata2017Q2=path+filename2017Q2
traindata2017Q3=path+filename2017Q3
traindata2017Q4=path+filename2017Q4
```

```python
train2017Q1 = pd.read_csv(traindata2017Q1,sep = ',', encoding="utf8")
train2017Q2 = pd.read_csv(traindata2017Q2,sep = ',', encoding="utf8")
train2017Q3 = pd.read_csv(traindata2017Q3,sep = ',', encoding="utf8")
train2017Q4 = pd.read_csv(traindata2017Q4,sep = ',', encoding="utf8")
```

```python
#Create table

FY2017=pd.concat([train2017Q1,train2017Q2,train2017Q3,train2017Q4], ignore_index=True)
FY2017
```

```python
FY2017.to_csv('FY2017.csv',";", index = False)
```
